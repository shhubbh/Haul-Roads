```python
import geopandas as gpd
from shapely.geometry import LineString, MultiLineString, Polygon
from shapely.ops import linemerge, unary_union
from skimage.morphology import skeletonize
import sknw
import numpy as np
import rasterio
from rasterio.features import rasterize
from scipy.spatial import Voronoi

def merge_lines(lines):
    # Merge lines and handle both LineString and MultiLineString outputs
    merged = linemerge(lines)
    if isinstance(merged, LineString):
        return [merged]
    elif isinstance(merged, MultiLineString):
        return list(merged.geoms)
    return []

def get_centerline(polygon, resolution=1):
    # Create a binary image from the polygon
    bounds = polygon.bounds
    transform = rasterio.transform.from_bounds(*bounds, int((bounds[2]-bounds[0])/resolution), int((bounds[3]-bounds[1])/resolution))
    binary_image = rasterize([polygon], out_shape=(int((bounds[3]-bounds[1])/resolution), int((bounds[2]-bounds[0])/resolution)), transform=transform, fill=0, all_touched=True)

    # Skeletonize the binary image
    skeleton = skeletonize(binary_image)

    # Generate centerlines using sknw
    graph = sknw.build_sknw(skeleton)
    
    # Convert the centerlines to LineString geometries
    lines = []
    for (s, e) in graph.edges():
        ps = graph[s][e]['pts']
        # Transform pixel coords to spatial coords, accounting for flipping
        coords = [(p[1] * resolution + bounds[0], (binary_image.shape[0] - p[0]) * resolution + bounds[1]) for p in ps]
        line = LineString(coords)
        lines.append(line)
    
    return lines

def voronoi_centerline(polygon):
    # Get the exterior coordinates of the polygon
    points = np.array(polygon.exterior.coords)
    vor = Voronoi(points)
    
    # Extract Voronoi ridge vertices
    lines = []
    for ridge in vor.ridge_vertices:
        if -1 not in ridge:  # -1 indicates vertex at infinity
            line = LineString(vor.vertices[ridge])
            lines.append(line)
    
    # Filter lines to keep only those inside the polygon
    filtered_lines = [line.intersection(polygon) for line in lines if polygon.intersects(line)]
    filtered_lines = [line for line in filtered_lines if not line.is_empty]
    
    return merge_lines(filtered_lines)

# Step 1: Read the shapefile
input_shapefile = r"C:\Users\shubh\Documents\Analytics\Haul Roads WS\HaulRoads Data Files\HaulRoads_SHP.shp"
gdf = gpd.read_file(input_shapefile)

# Step 2: Initialize a list to hold final centerlines
final_centerlines = []

# Step 3: Process each polygon to generate and merge centerlines
for poly in gdf.geometry:
    if isinstance(poly, Polygon):
        # Generate initial centerlines
        initial_lines = get_centerline(poly)
        merged_initial_lines = merge_lines(initial_lines)
        
        # Generate voronoi centerlines for better centering
        voronoi_lines = voronoi_centerline(poly)
        merged_voronoi_lines = merge_lines(voronoi_lines)
        
        # Combine and merge both sets of centerlines
        combined_lines = merged_initial_lines + merged_voronoi_lines
        merged_centerlines = merge_lines(combined_lines)
        
        final_centerlines.extend(merged_centerlines)
    else:
        print(f"Non-polygon geometry found: {type(poly)}. Skipping...")

# Step 4: Create a GeoDataFrame for the final centerlines
centerline_gdf = gpd.GeoDataFrame(geometry=final_centerlines, crs=gdf.crs)

# Step 5: Save the final centerlines to a new shapefile
output_shapefile = r"C:\Users\shubh\Documents\Analytics\Haul Roads WS\HaulRoads Data Files\HaulRoadsCL_corrected_SHP.shp"
centerline_gdf.to_file(output_shapefile)

print(f"Final centerlines have been saved to {output_shapefile}")
```

```python
import geopandas as gpd
from shapely.geometry import LineString, Polygon, MultiLineString
from shapely.ops import linemerge
from shapely.validation import make_valid
from skimage.morphology import skeletonize, thin
from scipy.ndimage import gaussian_filter
import sknw
import numpy as np
import rasterio
from rasterio.features import rasterize
import matplotlib.pyplot as plt

def merge_lines(lines):
    # Merge lines and handle both LineString and MultiLineString outputs
    merged = linemerge(lines)
    if isinstance(merged, LineString):
        return [merged]
    elif isinstance(merged, MultiLineString):
        return list(merged.geoms)
    return []

def smooth_line(line, smoothing_factor=3):
    x, y = line.xy
    x = gaussian_filter(x, smoothing_factor)
    y = gaussian_filter(y, smoothing_factor)
    return LineString(np.column_stack([x, y]))

def get_centerline(polygon, resolution=1):
    # Create a binary image from the polygon
    bounds = polygon.bounds
    transform = rasterio.transform.from_bounds(*bounds, int((bounds[2]-bounds[0])/resolution), int((bounds[3]-bounds[1])/resolution))
    binary_image = rasterize([polygon], out_shape=(int((bounds[3]-bounds[1])/resolution), int((bounds[2]-bounds[0])/resolution)), transform=transform, fill=0, all_touched=True)

    # Smooth the binary image
    binary_image = gaussian_filter(binary_image.astype(float), sigma=1)
    binary_image = (binary_image > 0.5).astype(np.uint8)

    # Skeletonize the binary image
    skeleton = thin(binary_image)

    # Generate centerlines using sknw
    graph = sknw.build_sknw(skeleton)
    
    # Extract the main branches from the skeleton
    lines = []
    for (s, e) in graph.edges():
        ps = graph[s][e]['pts']
        # Transform pixel coords to spatial coords, accounting for flipping
        coords = [(p[1] * resolution + bounds[0], (binary_image.shape[0] - p[0]) * resolution + bounds[1]) for p in ps]
        line = LineString(coords)
        lines.append(line)
    
    return lines

input_shapefile = r"C:\Users\shubh\Documents\Analytics\Haul Roads WS\HaulRoads Data Files\HaulRoads_SHP.shp"
gdf = gpd.read_file(input_shapefile)

final_centerlines = []

# generating centerlines by processing each polygon
for poly in gdf.geometry:
    if isinstance(poly, Polygon):
        # Generate initial centerlines
        initial_lines = get_centerline(poly)
        merged_initial_lines = merge_lines(initial_lines)
        
        # Validate and fix centerlines to avoid topology errors
        valid_centerlines = [make_valid(line) for line in merged_initial_lines]
        
        # Smooth the centerlines
        smoothed_centerlines = [smooth_line(line) for line in valid_centerlines]
        
        final_centerlines.extend(smoothed_centerlines)
    else:
        print(f"Non-polygon geometry found: {type(poly)}. Skipping...")

centerline_gdf = gpd.GeoDataFrame(geometry=final_centerlines, crs=gdf.crs)

output_shapefile = r"C:\Users\shubh\Documents\Analytics\Haul Roads WS\HaulRoads Data Files\HRCL_Final_SHP.shp"
centerline_gdf.to_file(output_shapefile)

print(f"Final centerlines have been saved to {output_shapefile}")


fig, ax = plt.subplots(figsize=(10, 10))
gdf.boundary.plot(ax=ax, linewidth=1, edgecolor='blue')
centerline_gdf.plot(ax=ax, linewidth=1, color='red')
plt.title("Polygons and Generated Centerlines")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
#plt.show()
```
